# Считываем нашу таблицу 

```python
import pandas as pd
import numpy as np
xl = pd.ExcelFile('Kofe11.xlsx')
# 1 Лист таблицы
df1 = xl.parse('Лист1')
list1 = np.array(df1[3:][['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3']])


# 2 Лист таблицы
df2 = xl.parse('Лист2')
list2 = np.array(df2[3:][['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3']])


# 3 Лист таблицы
df3 = xl.parse('Лист3')
list3 = np.array(df3[3:][['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3']])
```

# Склеиваем все полученные данные
Склеим все до этого полученные данные

```python
result = np.vstack((list1,list2,list3))
list(map(lambda i: [i[0],i[1]*100/i[2]], result))
pd.DataFrame(result, columns = ['Бренд','Цена','Вес'])
```

# Для дальнейшего анализа нам потребуется список всех брендов и их количество

```python
brend_count = list(map(lambda i: i[0], result))
count = dict(map(lambda x: [x,list(brend_count).count(x)] , brend_count))# Лямбда функция считывающая количество каждого бренда
print(brend_count)
print(count)
```

# С помощью Лямбда функций вычисляем цену каждого товара за 100 грамм

```python

brend = dict(map(lambda i: [i[0],[]], result)) # создаем словарь состоящий из всех брендов, со значениями пустых списков

Dict = list(map(lambda y: brend.setdefault(y[0]).append(y[1]),result)) # заполняем значения словаря
"""
При помощи Лямбда функции записали в result, в виде вложенных списков, бренды и их среднюю цену
"""
result = list(map(lambda i_key, i_value, j_value: [i_key,sum(i_value)/j_value] ,brend.keys(),brend.values(), count.values()))
print(result) 
```

# С помощью списка цен, находим суммую min и max, делим данное значение на количество сегментон(4).
# Благодаря этому методу мы полуили шаг сегмента, то есть [0,cut], [cut,2*cut] и тд, с помощью этого мы поделили все наши данные на равные ценовые сегменты

```python
result_price = list(map(lambda i: i[1], result)) # список цен
result_price = np.array(result_price) # массив цен

cut = float((result_price.min()+result_price.max())/4)
print("Шаг, с которым меняется сегмент =", cut)

"""
При помощи np.select заменим наши цены на сегменты (1,2,3,4), к которым принадлежат бренды.
1 сегмент = low
2 сегмент = middle
3 сегмент = upper middle
4 сегмент = high

"""
condlist = [result_price<cut, result_price<2*cut ,result_price<3*cut, 3*cut<=result_price]
choicelist = [1, 2, 3, 4]
result_segment = np.select(condlist, choicelist)
result_excel = list(map(lambda i,j: [i[0],j],result,result_segment))
print(result_excel)
```

```python
df = pd.DataFrame (result_excel,  columns=['Бренд', 'Сегмент'])
filepath = 'my_excel_file.xlsx'
df.to_excel(filepath)
df
```
